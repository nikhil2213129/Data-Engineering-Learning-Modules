```python
import random

random.seed(42)
_ALL_DATA = [{'id': i, 'value': round(random.uniform(1, 100), 2)} for i in range(100)]

def paginated_api(page=1, per_page=10):
    """Simulates a paginated API — DO NOT MODIFY"""
    start = (page - 1) * per_page
    end = start + per_page
    chunk = _ALL_DATA[start:end]
    return {'data': chunk, 'page': page, 'per_page': per_page,
            'total': len(_ALL_DATA), 'has_next': end < len(_ALL_DATA)}

print("Paginated API ready. Example:")
print(paginated_api(page=1))
```

```python
# Initialize variables
collected = []
page = 1
total_pages = 0

# Pagination loop
while True:
    response = paginated_api(page=page)
    
    # Collect data
    collected.extend(response['data'])
    
    # Track pages
    total_pages += 1
    
    # Check if more pages exist
    if not response['has_next']:
        break
    
    page += 1

print(f"Total pages fetched: {total_pages}")
print(f"Total records collected: {len(collected)}")
```

```python
assert len(collected) == 100, f"Expected 100 records, got {len(collected)}"
assert collected[0]['id'] == 0, "First record should be id=0"
assert collected[-1]['id'] == 99, "Last record should be id=99"
print(f"Collected all {len(collected)} records — all checks passed!")
```
