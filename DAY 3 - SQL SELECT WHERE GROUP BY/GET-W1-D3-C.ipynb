```python
import sqlite3, pandas as pd, random
from datetime import date, timedelta

random.seed(42)
conn = sqlite3.connect(':memory:')
conn.execute('CREATE TABLE sales (id INT, product TEXT, region TEXT, qty INT, price REAL, sale_date TEXT)')
for i in range(100):
    conn.execute('INSERT INTO sales VALUES (?,?,?,?,?,?)',
        (i, random.choice(['Laptop','Monitor','Keyboard','Mouse','Headset']),
         random.choice(['North','South','East','West']),
         random.randint(1, 10), round(random.uniform(10, 500), 2),
         str(date(2024,1,1) + timedelta(days=random.randint(0, 180)))))
conn.commit()
print("Database ready: 100 rows in 'sales' table")
pd.read_sql("SELECT * FROM sales LIMIT 5", conn)
```

```python
q1 = """SELECT region, SUM(qty * price) AS revenue
FROM sales
GROUP BY region
ORDER BY revenue DESC"""

q1 = pd.read_sql(q1,conn)
print(q1)
```

```python
q2 = """SELECT product, SUM(qty) AS total_qty
FROM sales
GROUP BY product
ORDER BY total_qty DESC
LIMIT 3"""
q2 = pd.read_sql(q2,conn)
print(q2)
```

```python
q3 = """SELECT COUNT(*) AS count
FROM sales
WHERE price > 200"""
q3 = pd.read_sql(q3,conn)
print(q3)
```

```python
q4 = """SELECT substr(sale_date, 1, 7) AS month, COUNT(*) AS order_count
FROM sales
GROUP BY month
ORDER BY month"""
q4 = pd.read_sql(q4,conn)
print(q4)
```

```python
q5 = """SELECT region, product, COUNT(*) AS order_count
FROM sales
GROUP BY region, product
ORDER BY region, product"""
q5 = pd.read_sql(q5,conn)
print(q5)
```
