Step 1: Data Setup 

```python
import pandas as pd, random
from datetime import date,timedelta

random.seed(42)
flat = pd.DataFrame([{
    'order_id': f'O-{i:03d}',
    'product': random.choice(['Laptop','Mouse','Monitor']),
    'category': random.choice(['Electronics','Accessories']),
    'region': random.choice(['North','South','East','West']),
    'qty': random.randint(1, 5),
    'price': round(random.uniform(10, 500), 2),
    'order_date': str(date(2024,1,1) + timedelta(days=random.randint(0, 180)))
} for i in range(100)])
flat.head()
```

```python
dim_product = flat[['product','category']].drop_duplicates().reset_index(drop=True)
dim_product.index.name = 'product_id'
dim_product = dim_product.reset_index()

dim_region = pd.DataFrame({'region': flat['region'].unique()})
dim_region.index.name = 'region_id'
dim_region = dim_region.reset_index()

fact_sales = flat.merge(dim_product, on=['product','category'], how='left')
fact_sales = fact_sales.merge(dim_region, on='region', how='left')
fact_sales['revenue'] = fact_sales['qty'] * fact_sales['price']
fact_sales = fact_sales[['order_id','product_id','region_id','qty','price','order_date','revenue']]
```

```python
# Run this cell to validate your solution
assert len(fact_sales) == 100, "Fact table should have 100 rows"
assert 'product_id' in fact_sales.columns, "Missing product_id"
assert 'region_id' in fact_sales.columns, "Missing region_id"
assert len(dim_product) <= 6, "Too many product dimension rows"
assert len(dim_region) <= 4, "Too many region dimension rows"
print("Star schema — all checks passed!")
```
