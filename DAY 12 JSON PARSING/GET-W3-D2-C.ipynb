```python
import json, pandas as pd, random
random.seed(42)

nested = [
    {
        'user': {
            'id': i,
            'name': f'User_{i}',
            'address': {
                'city': random.choice(['NYC','LA','Chicago']),
                'zip': str(random.randint(10000,99999))
            }
        },
        'orders': [
            {
                'item': random.choice(['Book','Pen','Bag']),
                'price': round(random.uniform(5,50),2)
            }
            for _ in range(random.randint(1, 3))
        ]
    }
    for i in range(20)
]

print(f"nested json: {len(nested)} users")
print(json.dumps(nested[0], indent=2))

# Flatten
flatten = []
for entry in nested:
    user = entry.get('user', {})
    orders = entry.get('orders', [])
    for order in orders:
        flatten.append({
            'user_id': user.get('id', 'Unknown'),
            'name': user.get('name', 'Unknown'),
            'city': user.get('address', {}).get('city', 'Unknown'),
            'zip': user.get('address', {}).get('zip', 'Unknown'),
            'item': order.get('item', 'Unknown'),
            'price': order.get('price', 'Unknown')
        })

df = pd.DataFrame(flatten)
print(df.head())
```

```python
# Run this cell to validate your solution
assert len(df) > 20, "Should have more than 20 rows (users have multiple orders)"
assert set(df.columns) >= {'user_id','name','city','item','price'}, "Missing columns"
assert df['user_id'].nunique() == 20, "Should have 20 unique users"
assert df['price'].sum() > 0, "Prices should sum > 0"
print("All checks passed!")
```
