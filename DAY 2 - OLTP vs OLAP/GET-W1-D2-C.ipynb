```python
import pandas as pd, random
from datetime import date, timedelta

random.seed(42)
rows = [{'txn_id': f'T-{i:03d}',
         'product': random.choice(['Laptop','Monitor','Keyboard']),
         'region': random.choice(['North','South','East','West']),
         'amount': round(random.uniform(10, 500), 2),
         'txn_date': str(date(2024,1,1) + timedelta(days=random.randint(0, 180)))}
        for i in range(100)]
df = pd.DataFrame(rows)
print(f"Transactions: {len(df)} rows")
df.head()
```

```python
olap = df.groupby(['product', 'region']).agg(
    revenue=('amount', 'sum'),
    order_count=('txn_id', 'count'),
    avg_order=('amount', 'mean')
).reset_index()

olap['revenue'] = olap['revenue'].round(2)
olap['avg_order'] = olap['avg_order'].round(2)
```

```python
print("OLAP Summary Table:")
display(olap)
```
